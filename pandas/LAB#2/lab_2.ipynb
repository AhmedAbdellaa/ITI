## Question#3:

```python
#read csv file
Telco_df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
Telco_df.info()
Telco_df.head()
```

```python
Telco_df.set_index('customerID' ,inplace=True) #convert id to record index
Telco_df.drop_duplicates(inplace=True) # drop duplicate
Telco_df.info()
```

```python
#convert TotalCharges type to float as it has numeric values
Telco_df.TotalCharges =pd.to_numeric(Telco_df['TotalCharges'] , errors = 'coerce') 
Telco_df.info()
```

```python
#fill missed data of TotalCharges with average of it's column
Telco_df.fillna(value={'TotalCharges':Telco_df['TotalCharges'].mean()} , inplace=True)
Telco_df.info()
```

```python
#handle missing values
Telco_df.dropna(inplace = True)
Telco_df.info()
```

```python
#get list of categorical data ,numerical data to use it in normalization and one hot encoding 
Telco_df_cols_cat = []
Telco_df_cols_num = []
for col in Telco_df.head() :
    if Telco_df[col].dtype == 'object' :
        if not (col =='Contract' or col == 'Churn'):
            Telco_df_cols_cat.append(col)
    elif Telco_df[col].dtype == 'float64' :
        Telco_df_cols_num.append(col)

Telco_df_cols_num.append('tenure')
print(Telco_df_cols_cat)
print(Telco_df_cols_num)
```

```python
#one hot encoding for categorical
Telco_df = pd.concat([Telco_df, pd.get_dummies(Telco_df[Telco_df_cols_cat])], axis=1)
for idx in Telco_df_cols_cat :
    del Telco_df[idx]

#order encoding
Contract_ordinal_encoding = {"Month-to-month":1, "One year":2, "Two year":3}
Telco_df["Contract"] = Telco_df["Contract"].replace(Contract_ordinal_encoding)

Churn_ordinal_encoding = {"No": 0 ,"Yes" : 1 }
Telco_df["Churn"] = Telco_df["Churn"].replace(Churn_ordinal_encoding)

#normalize numerical columns
#numerical_norm = ['MonthlyCharges' , 'TotalCharges' ,'tenure'  ]
for n in Telco_df_cols_num:
     Telco_df[n]= (Telco_df[n]-Telco_df[n].min()) / (Telco_df[n].max()-Telco_df[n].min())
    
Telco_df.info()
Telco_df.head()
```

```python
#split target label for feature train
target_df = pd.DataFrame(Telco_df['Churn'] )
del Telco_df['Churn']

target_df.info()

target_df.to_csv('target_df.csv')
Telco_df.to_csv('Telco_df.csv')
```
